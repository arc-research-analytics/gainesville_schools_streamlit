```python
import os
import pandas as pd
from functools import reduce
import glob
import geopandas as gpd
```

```python
path1 = 'Tabular/Esri/' 

files = glob.glob(os.path.join(path1, "*.csv"))

for file in files:
    df = pd.read_csv(file)
    df['Block Group'] = df['Block Group'].astype(str)

df = reduce(lambda l,r: l.merge(r, on='Block Group'), [pd.read_csv(f) for f in files])

df.head()
```

```python
df['Block Group'] = df['Block Group'].astype(str)
df['Block Group'] = df['Block Group'].str.replace('.','')

df.head()
```

```python
# df.to_csv('Gainesville_test.csv')
gdf = gpd.read_file('Geospatial/Gainesville_BG.gpkg')
gdf = gdf[['GEOID', 'geometry']]
gdf.head()
```

```python
new_gdf = gdf.merge(df, left_on = 'GEOID', right_on = 'Block Group')
new_gdf.columns
# new_gdf.to_file('master_file.gpkg')
```
